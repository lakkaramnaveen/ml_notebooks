```python
import requests
from bs4 import BeautifulSoup

url = 'https://benscott.dev/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

links = []
for link in soup.find_all('a'):
    href = link.get('href')
    if href and href.startswith('http'):
        links.append(href)

links[:10]
```

```python
import re
import pandas as pd

def tokenize_url(url):
    tokens = re.split('/|-|_|\.|\?', url.lower())
    return [token for token in tokens if token]

tokens_list = [tokenize_url(link) for link in links]
df_tokens = pd.DataFrame(tokens_list, columns=[f"Token {i+1}" for i in range(len(max(tokens_list, key=len)))])

df_tokens.head()
```

```python
import requests
import time

urls = ['https://benscott.dev/', 'https://benscott.dev/#projects']

for url in urls:
    response = requests.get(url)
    print(f"Status Code for {url}: {response.status_code}")
    time.sleep(1)
```
